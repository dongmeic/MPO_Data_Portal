```python
def get_blockIDs(citynm):
    # require city and block data
    print(citynm)
    selected_city = arcpy.management.SelectLayerByAttribute(city, "NEW_SELECTION", "FIRST_NAME = '{0}'".format(citynm), None)
    selected_blocks = arcpy.management.SelectLayerByLocation(tabblock10, "HAVE_THEIR_CENTER_IN", selected_city, None, "NEW_SELECTION", "NOT_INVERT")
    layer_object = selected_blocks.getOutput(0)
    blockIDs = [row[0] for row in   arcpy.da.SearchCursor(layer_object, "BLOCKID10")]
    return(blockIDs)
```

```python
blockIDs = get_blockIDs(citynm = 'Adams')
```

```python
cities['BlockIDs'] = cities.City.apply(lambda x: get_blockIDs(x))
```
